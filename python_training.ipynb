# Dictionaries

```python
dictWithBrackets = {'Nombre1': 'Valor1', 'Nombre2': 'Valor2', 'Nombre3': 3}
print(dictWithBrackets)

dictWithDict = dict(Nombre4 = 'Valor4', Nombre5 = 5, Nombre6 = 6.)
print(dictWithDict)
print(type(dictWithBrackets))
print(type(dictWithDict))
```

```python
# Para extraer valores, se hace en forma de arbol, es decir, para extraer "Valor de prueba 2" 
# se selecciona por [], en este caso sería dictMixto['Nombre6']['prueba2']

dictMixto = dict(Nombre4 = 'Valor4', Nombre5 = 5, Nombre6 = dict(prueba1 = 2, prueba2 = 'Valor de prueba 2'))
dictMixto
```

```python
# Extrae el valor en Nombre4
dictMixto['Nombre4']
```

```python
# Esto generara un error ya que no existe ningun valor llamado prueba3 en la sección Nombre6
dictMixto['Nombre6']['prueba3']
```

```python
var1 = 'Nombre6'
var2 = 'prueba2'
dictMixto[var1][var2]
```

```python
# Muestra información de la lectura
datos
```

```python
# Muestra tipo de datos
type(datos)
```

```python
# Genera un primer analisis general del contenido del dataframe
datos.describe()
```

```python
# Muestra el tipo de datos contenidos 
datos.info()
```

```python
# Análisis sobre columnas seleccionadas
datos[['sexo', 'edad', 'entidad']].describe()
```

```python
# Tipo de datos sobre columnas seleccionadas
datos[['sexo', 'edad', 'entidad']].info()
```

```python
# Se pueden agregar datos por columnas de manera rapida
datos['edad'].sum()
```

```python
datos[['sexo', 'edad', 'entidad']].median()
```

```python
# Para poder filtrar se pueden dar condiciones, como ejemplo que entidad sea = a 10
print((datos['entidad'] == 10))
# Cual es el valor del resultado de comparar entidad con el valor 10 en la posición 4
print((datos['entidad'] == 10)[3])
# Tipo de dato resultado de comparar entidad con el valor 10 en la posición 4
type((datos['entidad'] == 10)[3])
```

```python
# Valores unicos de la columna
datos['edad'].unique()

datos['edad'].describe()
```

```python
# Las comparaciones son:
# == iguales | != diferentes | <, >, <=, >= rango
#datos.query('edad == 29')
#datos.query('edad >= 29')
datos.query('edad == 17')
```

```python
datos[datos['edad']==35]
```

```python
# Filtrar datos de manera sencilla y explicita
datos.query('entidad == 9 and sexo == 2')
```

```python
# Para graficar existen multiples librerias en este caso se usa plotly
import plotly.express as px

fig = px.histogram(x = datos.query('edad >= 17 and edad <= 30 and facultad == 11')['facultad'])

fig.show()
```

```python
fig = px.scatter(datos, x = datos['entidad'], y = datos['edad'])
fig.show()
```

```python
# Agrupar datos por condiciones
datos.groupby(['entidad', 'sexo']).agg(ConteoSexo=('sexo', 'count'), ConteoEntidad = ('entidad', 'count')).reset_index()
```
